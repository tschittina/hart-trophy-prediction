```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/skaters.csv')
df = df.drop(columns=['Unnamed: 0'])
df.head(8)
```

```python
df1 = pd.read_csv('data/goalies.csv')
df1 = df1.drop(columns=['Unnamed: 0'])
df1.head(8)
```

```python
fix, axs = plt.subplots(nrows=2, ncols=2, figsize = (12,8))
axs[0,0].scatter(df['ops'], df['vote-share'])
axs[0,0].set_title('OPS (x) vs. Vote Share (y)')

axs[0,1].scatter(df['dps'], df['vote-share'])
axs[0,1].set_title('DPS (x) vs. Vote Share (y)')

axs[1,0].scatter(df['tps'], df['vote-share'])
axs[1,0].set_title('TPS (x) vs. Vote Share (y)')

axs[1,1].scatter(df['plus-minus'], df['vote-share'])
axs[1,1].set_title('+/- (x) vs. Vote Share (y)')

plt.show()
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12,8))

axs[0,0].scatter(df['goals'], df['vote-share'])
axs[0,0].set_title('Goals (x) vs. Vote Share (y)')

axs[0,1].scatter(df['assists'], df['vote-share'])
axs[0,1].set_title('Assists (x) vs. Vote Share (y)')

axs[1,0].scatter(df['points'], df['vote-share'])
axs[1,0].set_title('Points (x) vs. Vote Share (y)')

plt.show()
```

```python
counts = []
for pos in ['C', 'W', 'D']:
    num = df[df[f'position_{pos}'] == 1]['mvp'].sum()
    counts.append(num)
    
counts.append(df1['mvp'].sum())
counts
```

```python
positions = ['Center', 'Wing', 'Defense', 'Goalie']
title = "MVP Awards Won by Position (1980 - 2024)"

fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(positions, height=counts)
ax.set_title(title)
plt.show()
```
