```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/skaters.csv')
df = df.drop(columns=['Unnamed: 0'])
df
```

```python
fix, axs = plt.subplots(nrows=2, ncols=2, figsize = (12,8))
axs[0,0].scatter(df['ops'], df['vote-share'])
axs[0,0].set_title('OPS (x) vs. VS (y)')

axs[0,1].scatter(df['dps'], df['vote-share'])
axs[0,1].set_title('DPS (x) vs. VS (y)')

axs[1,0].scatter(df['tps'], df['vote-share'])
axs[1,0].set_title('TPS (x) vs. VS (y)')

axs[1,1].scatter(df['plus-minus'], df['vote-share'])
axs[1,1].set_title('+/- (x) vs. VS (y)')

plt.show()
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12,8))

axs[0,0].scatter(df['goals'], df['vote-share'])
axs[0,0].set_title('Goals (x) vs. VS (y)')

axs[0,1].scatter(df['assists'], df['vote-share'])
axs[0,1].set_title('Goals (x) vs. VS (y)')

axs[1,0].scatter(df['points'], df['vote-share'])
axs[1,0].set_title('Points (x) vs. VS (y)')

plt.show()
```
